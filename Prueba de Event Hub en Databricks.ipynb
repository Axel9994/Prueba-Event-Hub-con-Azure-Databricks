# Inicio

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
#Eventhub namespace and required keys
EH_NAMESPACE_NAME = "[TU_EVENT_HUB]"
EH_KEY_NAME = "[NOMBRE_LLAVE_EVENT_HUB]"
EH_KEY_VALUE = "[LLAVE_EVENT_HUB]"
# connection string of Event Hubs Namespace
connectionString = "Endpoint=sb://{0}.servicebus.windows.net/;SharedAccessKeyName={1};SharedAccessKey={2};EntityPath=kafkaenabledhub".format(EH_NAMESPACE_NAME, EH_KEY_NAME, EH_KEY_VALUE)
ehConfig = {}
ehConfig['eventhubs.connectionString'] = connectionString
ehConfig['eventhubs.connectionString'] = sc._jvm.org.apache.spark.eventhubs.EventHubsUtils.encrypt(connectionString)
EventHubdf = spark.readStream.format("eventhubs").options(**ehConfig).load()
```

Verificacion de Event Hub DataFrame

```python
print( EventHubdf.isStreaming)
print( EventHubdf.printSchema())
```

Crear Esquema Json

```python
jsonschema = StructType() \
.add("id", StringType()) \
.add("timestamp", TimestampType()) \
.add("rpm", IntegerType()) \
.add("speed", IntegerType()) \
.add("kms", IntegerType())
```

### Transformaciones 

```python
newEventHubDf=EventHubdf.selectExpr("CAST(body AS STRING)")
```

```python
display(newEventHubDf)
```

```python
newEventHubDf=newEventHubDf.withColumn('vehiclejson', from_json(col('body'),schema=jsonschema))
```

```python
display(newEventHubDf)
```

```python
newEventHubjsonDF=newEventHubDf.select("body", "vehiclejson.*")
```

```python
display(newEventHubjsonDF)
```

```python
newEventHubjsonDF.selectExpr(
    "id" \
    ,"timestamp" \
    ,"rpm" \
    ,"speed" \
    ,"kms" ) \
.writeStream.format("delta") \
.outputMode("append") \
.option("checkpointLocation", "dbfs:/Vehiclechkpoint_AzureEventHub_Demo/") \
.option("mergeSchema", "true") \
.start("dbfs:/Vehiclechkpoint_AzureEventHub_Delta")
```

```python
display(newEventHubjsonDF)
```

```python
new_df = spark.readStream \
    .format("delta") \
    .load("dbfs:/Vehiclechkpoint_AzureEventHub_Delta") \
    .select("id", "timestamp", "rpm", "speed", "kms")
```

```python
display(new_df)
```

```python
avg_stats_df = new_df.groupBy(year("timestamp"), month("timestamp"), day("timestamp"), hour("timestamp"), minute("timestamp")) \
    .agg(avg("rpm"), avg("speed") , avg("kms")) \
    .withColumnRenamed("year(timestamp)", "year") \
    .withColumnRenamed("month(timestamp)", "month") \
    .withColumnRenamed("dayofmonth(timestamp)", "day") \
    .withColumnRenamed("hour(timestamp)", "hour") \
    .withColumnRenamed("minute(timestamp)", "minute")
```

```python
display(avg_stats_df)
```
